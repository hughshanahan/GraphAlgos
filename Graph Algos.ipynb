```python
# PageRank
import numpy as np

def findPageRank( PT, c, epsilon):
# PT - Transpose of P, c - coefficient, epsilon - cutoff
        n = PT.shape[0]
        delta = 10.0 * epsilon
        v = np.full((n),1.0/n)
        # Create vector         of n entries, each entry with 1/n
        x = np.full((n),1.0/n)
        # assume ranks are all the same initially
        y = np.zeros((n))
        while delta > epsilon:
                y = c * np.dot(PT, x)
                gamma = L1(x) - L1(y)
                # L1 computes L1 norm
                y += gamma * v
                delta = L1( y - x )
                x = y.copy()
        return(x)

def L1(x):
    return np.linalg.norm(x,ord=1)

```
